```python
#File Name: PathakHw1B
#Author: Shivam Pathak
#Date: 2.8.25
#Complete Programs + Extra credit
#Question 1: Number Guessing Game using Binary Search Algorithm

import random


def binary_search(arr, target):
    """Function that conducts binary search"""
    left, right = 0, len(arr) - 1
    guess_count = 0  # Initialize guess count

    while left <= right:
        guess_count += 1  # Increment guess count
        mid = (left + right) // 2  # Find the middle index

        if arr[mid] == target:
            return guess_count  # Return the number of guesses
        elif arr[mid] < target:
            left = mid + 1  # Search the right half
        else:
            right = mid - 1  # Search the left half

    return guess_count  # Should never reach this if target is in range

def run_exp(max_number, trials=10_000):
    """Function that runs experiment"""
    total_guess_count = 0  # Counter for total guesses
    arr = list(range(1, max_number + 1))  # List from 1 to max_number

    for i in range(trials):
        target = random.randint(1, max_number)  # Generate a random number within the correct range
        total_guess_count += binary_search(arr, target)  # Call binary search and count guesses

    avg_guesses = total_guess_count / trials  # Compute the average guesses
    return total_guess_count, avg_guesses

# Run the experiments with correct ranges
total_1k, avg_1k = run_exp(1_000)  # Range 1k
total_1m, avg_1m = run_exp(1_000_000)  # Range 1M

# Print results
print(f"1. The random numbers between 1 .. 1K: Total guesses: {total_1k}, Avg: {avg_1k}")
print(f"2. The random numbers between 1 .. 1M: Total guesses: {total_1m}, Avg: {avg_1m}")



```

```python
#Question 2 Timer function

import time

#Step 2: Number Function (Taken from Online Guidlines  allow)
def listPrimeNumbers(theMaxNum):
    """Finds all prime numbers from 0 to theMaxNum using the Sieve of Eratosthenes."""
    if theMaxNum < 2:
        return []  # No primes if theMaxNum is less than 2

    primes = [True] * (theMaxNum + 1)  # Assume all numbers are prime
    primes[0], primes[1] = False, False  # 0 and 1 are NOT prime numbers

    for num in range(2, int(theMaxNum ** 0.5) + 1):  # Only check up to sqrt(theMaxNum)
        if primes[num]:  # If number is prime
            for multiple in range(num * num, theMaxNum + 1, num):  # Mark multiples as not prime
                primes[multiple] = False

    # Extract all prime numbers
    prime_list = [num for num, is_prime in enumerate(primes) if is_prime]

    return prime_list

#Step 1 Time Effencieny
import time

# Step 1: Time Efficiency Function

def timeEfficiency(funcName, *args, **kwargs):
    """Measures execution time of a function."""
    # Normalize the base time
    start = time.time()   # Start time relative to base
    result = funcName(*args, **kwargs)  # Run the function
    end = time.time()  # End time relative to base

    total_time = end - start  # Calculate total time

    # Store results in a dictionary
    result_dict = {
        'start': round(start, 6),
        'end': round(end, 6),
        'time efficiency': round(total_time, 6),
        'result': result[:10] + ["..."] + result[-10:] if len(result) > 20 else result
    }

    # Print formatted output
    print("{")
    for key, value in result_dict.items():
        print(f"   '{key}': {value},")
    print("}")

    return result_dict  # Return the result


#Step 3 Test driver
def main():
    """Test Driver to measure the exeuction time of the Prime Number Function."""

    #User input
    theMaxNum = int(input("Enter a number for the list of prime numbers:"))

    #Time efficiency of listprimenumbers
    print(f"\nList of prime numbers of {theMaxNum}")
    timeEfficiency(listPrimeNumbers, theMaxNum)



#Run Test

if __name__ == "__main__":
    main()

```

```python
#Question 3: Generate Random Numbers

import random
from itertools import permutations

def generate_random_numbers():
    """Generates random numbers between 0 and 9 (order matters)"""
    return tuple(random.randint(0, 9) for _ in range(3))


#3A Deterministic Brute Force Guessing Algorithm

def brute_force_guess(target):
    """Brute-force method to guess 3 numbers in order"""

    guesses = 0

    #Generate all the permutations of (0-9) taken 3 at a time

    for perm in permutations(range(10), 3):
        guesses += 1
        if perm == target:
            return guesses #Return number of tries when found

    return guesses #Should always find the number


#Question 3B Pure Random Guessing Algorithm

def random_guess(target, max_attempts=10_000):
    """Randomly guesses numbers until it finds the correct set"""

    guesses = 0

    while guesses < max_attempts:
        guesses += 1
        guess = tuple(random.randint(0, 9) for _ in range(3))
        if guess == target:
            return guesses #Return number of tries when found

    return max_attempts #if it reaches max attempts, return max


def run_exp(num_tries=100_000):
    """Runs the deterministic and random guessing experiment."""

    #Brute Force
    brute_force_total = 0
    brute_force_high = 0
    brute_force_low = float('inf')

    #Random Guess
    random_total = 0
    random_high = 0
    random_low = float('inf')
    bailouts = 0  #Count cases that exceed max_attempts

    for i in range (num_tries):
        if i % 1000 == 0:
            print(f"Running trial {i}/{num_tries}...")
        target = generate_random_numbers()

        #Brute Force
        brute_attempts = brute_force_guess(target)
        brute_force_total += brute_attempts
        brute_force_high = max(brute_force_high, brute_attempts)

        if brute_attempts < brute_force_low and brute_attempts >= 3:
            brute_force_low = brute_attempts

        #Random Guess
        random_attempts = random_guess(target)
        random_total += random_attempts
        random_high = max(random_high, random_attempts)
        random_low = min(random_low, random_attempts)

        if random_attempts == 10_000:
            bailouts += 1 #Count how many times max was hit

    #Calc Averages
    brute_force_avg = brute_force_total / num_tries
    random_avg = random_total / num_tries

    #Result Display
    print("\na)Deterministic brute-force guessing algorithm")
    print(f"Number of tries: {num_tries}")
    print(f"The highest number of guesses in a try: {brute_force_high}")
    print(f"The lowest number of tries: {brute_force_low}")
    print(f"The average number of tries: {brute_force_avg}")

    print("\nb)Random guessing algorithm")
    print(f"Number of tries: {num_tries}")
    print(f"The highest number of guesses in a try: {random_high}")
    print(f"The lowest number of tries: {random_low}")
    print(f"The average number of tries: {random_avg}")
    print(f"Number of bailouts: {bailouts}")

if __name__== "__main__":
    run_exp(10_000) #Change to 100,000 for the final testing
```

```python
# Q4) Collatz 3n+1 Algorithm

import time

def collatz_sequence_length(n, cache):
    """Computes the Collatz length for n"""
    original_n = n
    steps = 0 #Track steps
    max_value = n #largest value in sequence

    sequence = [] #track numbers in sequence
    while n != 1:
        if n in cache:
            steps += cache[n]
            break

        sequence.append(n) #store current number within sequence

        if n % 2 == 0:
            n = n//2  # even
        else:
            n=3 * n + 1 #odd

        max_value = max(max_value, n) #update Max
        steps += 1
    for i, num in enumerate(sequence):
        cache[num] = steps - i #store correct length
    return steps,max_value


def analyze_collatz(limit):
    """Analyzes Collatz sequence for numbers up to 'limit' and finds the top ten longest sequences."""

    cache = {1: 1}  #Memoization cache
    top_sequences = []  #Store the sequence length number and max value

    for num in range(1, limit + 1):
        length, max_value = collatz_sequence_length(num, cache)
        top_sequences.append((length, num, max_value))

        if num % 100000 == 0: #Progress
            print(f"Processed: {num}/{limit}")

    #sort by the length of the sequence (descending) and get top 10
    top_sequences.sort(reverse=True, key=lambda x: x[0])
    return top_sequences[:10] #return top 10 results


def main():
    """Main function to run the analysis of the Collatz sequence."""

    start_time = time.time()
    limit = 1_000_000 #range is adjustable

    print(f"\nRunning Collatz Sequence Analysis for {limit} numbers...")

    top_results = analyze_collatz(limit)

    print("\n Top 10 Longest Collatz Sequences \n")

    for idx, (length, num, max_value) in enumerate(top_results, 1):
        print(f"{idx}. Collatz sequence for {num}: {length} steps, Largest value: {max_value}")

    print(f"\nExecution Time: {round(time.time() - start_time, 2)} seconds.")


if __name__ == "__main__":
    main()



```

```python
# Q5) Extra Credit Decorator

import time
from functools import wraps
#Step 1 Time Effencieny
def timeEfficiency(func):
    """Measures execution time of a function."""
    @wraps(func)
    def wrapper(*args, **kwargs):
        # Normalize the base time
        start = time.time()   # Start time relative to base
        result = func(*args, **kwargs)  # Run the function
        end = time.time()  # End time relative to base

        total_time = round(end - start, 6)  # Calculate total time
        if isinstance(result, list):
            formatted_result = result[:10] + ["..."] + result[-10:] if len(result) > 20 else result
        else:
            formatted_result = result


        # Print formatted output
        print("{")
        print(f"   'start': {round(start, 6)},")
        print(f"   'end': {round(end, 6)},")
        print(f"   'time efficiency': {total_time}")
        print(f"'result': {result[:10] + ["..."] + result[-10:] if len(result) > 20 else result}")
        print("}")
        return result  # Return the result
    return wrapper #returns decorated function

#Step 2: Number Function (Taken from Online Guidlines  allow)
@timeEfficiency #Applies decorator
def listPrimeNumbers(theMaxNum):
    """Finds all prime numbers from 0 to theMaxNum using the Sieve of Eratosthenes."""
    if theMaxNum < 2:
        return []  # No primes if theMaxNum is less than 2

    primes = [True] * (theMaxNum + 1)  # Assume all numbers are prime
    primes[0], primes[1] = False, False  # 0 and 1 are NOT prime numbers

    for num in range(2, int(theMaxNum ** 0.5) + 1):  # Only check up to sqrt(theMaxNum)
        if primes[num]:  # If number is prime
            for multiple in range(num * num, theMaxNum + 1, num):  # Mark multiples as not prime
                primes[multiple] = False

    # Extract all prime numbers
    prime_list = [num for num, is_prime in enumerate(primes) if is_prime]

    return prime_list



# Step 1: Time Efficiency Function



#Step 3 Test driver
def main():
    """Test Driver to measure the exeuction time of the Prime Number Function."""

    #User input
    theMaxNum = int(input("Enter a number for the list of prime numbers:"))

    #Time efficiency of listprimenumbers
    print(f"\nList of prime numbers of {theMaxNum}")
    result = listPrimeNumbers(theMaxNum)
    print(f"\nFound {len(result)} prime numbers.") #confirming



#Run Test

if __name__ == "__main__":
    main()


```

2)What are first-class objects? What is the significance of functions being first-class objects?
First class objects are entities that can be assigned to variables, passed as arguments, and returned from functions. In python specifically functions are first class objects , meaning they can be stored, modified and used dynamically. The significance of this is that they enable higher-order functions, support functional programming, and allow decorators and callbacks. An example would be a function being assigned to a variable and printed as the variable.


3)What are inner functions? Why are they important for decorators?
An inner function is a function that is defined within another function(like def_wraps from above). It remains purely local to the function it is inside and is not accessible by anything else. They are important for decorators because it provides encapsulation so they are not accessible outside of their function, making sure there is better control overall. They also retain access to the enclosed function's variables even after the function has been executed. An example of this is the def_wraps from above.
